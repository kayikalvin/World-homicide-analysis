```python
#filtering region[asia and europe] and year> 2016
df4 = df[df['Region'].isin(['Asia','Europe']) & (df.Year >  2016)].reset_index(drop=True)
df4
```

```python
df4= df4[['Region','Year','Count']]
df4
```

```python
y = df4.groupby(['Region','Year']).sum()['Count']
y
```

```python
#unstack in order to plot
df_unstacked = y.unstack(level=0)
df_unstacked
```

```python
df_unstacked.index = df_unstacked.index.astype(int).astype(str)
```

```python
# visualize
df_unstacked.plot(kind = 'line',figsize = (10,6))
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count of Asia and Europe over the Years')
plt.show()
```

```python
# group by subregion and find average count within each subregion and return treemap
df7 = df.groupby('Subregion')['Count'].mean().sort_values(ascending = False).round(2)
df7
```

```python
# visualize
df7.index
```

```python
df7.values
```

```python
data = {
    'Category' : df7.index,
    'Value' : df7.values,
    'info' : df7.values
}
df = pd.DataFrame(data)
df
```

```python
fig = px.treemap(df,path = ['Category'],value = 'Value',title = 'Treemap')
fig.update_traces(hovertemplate = 'Category:%{'label'}<br>Value:%{value}')
fig.show()
```
